```python
# Do not edit. These are the only imports permitted.
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
dataset = pd.read_csv('/Users/yingxuechen/Desktop/Golf-Player-Study/exploratory_data_analysis/EDA_Python/balanced_data.csv')
dataset
```

### map categorical variables to dummy variables

```python

dataset_new = dataset.copy()# exclude the label

for col_num, col_name in enumerate(dataset_new):
    col_data = dataset_new[col_name] # get data for each column
    col_dtype = col_data.dtype # get dtype for each column
    if col_dtype == 'object':
        dataset_new = dataset_new.drop(col_name, 1) # delete the column that needs mapping
        value_sets = col_data.unique() # to get the range of values
        for value_unique in value_sets:
            col_name_new = col_name + '_' + value_unique # create unique column which would use each value in the value_sets
            col_tmp = dataset.iloc[:, col_num] # get the original column
            new_col = (col_tmp == value_unique)
            dataset_new[col_name_new] = new_col

dataset_new
```

### To map boolean variables to int

```python
for u in dataset_new.columns:
    if dataset_new[u].dtype==bool:
        dataset_new[u]=dataset_new[u].astype('int')

print('num of observations: ', dataset_new.shape[0])
print('num of features: ', dataset_new.shape[1] - 2, '| exclude label and golferid which is used as identifier')
# active : inactive = 7.125 : 1
print('<is_active = TRUE> Vs <is_active = FALSE> = ', sum(dataset['is_active'] == 1), ':', sum(dataset['is_active'] == 0))
dataset_new
```

### split dataset into train and test

```python
y = dataset_new['is_active'].values
X = dataset_new.drop(columns=['is_active', 'golferid']).values
print(type(y), type(X))

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
print(X_train[0], y_train[0])
# dataset_new_arr = dataset_new_features.values
# n_train = round(dataset_new_arr.shape[0]*0.8)
# n_test = dataset_new_arr.shape[0] - n_train
# print('train | test:', n_train, ' | ', n_test)

# train_set, test_set = dataset_new_arr[:n_train], dataset_new_arr[n_train:]
# labels_train, labels_test = labels[:n_train], labels[n_train:]
# print(len(train_set), len(test_set))
# print(len(labels_train), len(labels_test))
# labels_train
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
svm_p = SVC(kernel='poly', gamma='scale')
svm_p.fit(X_train, y_train)
```

```python
svm_p0 = SVC(kernel='poly', gamma='scale', degree=2)
svm_p0.fit(X_train, y_train)
```

```python
svm_p1 = SVC(kernel='poly', gamma='scale', degree=4)
svm_p1.fit(X_train, y_train)
```

```python
svm_r = SVC(kernel='rbf', gamma='scale')
svm_r.fit(X_train, y_train)
```

```python
svm_r1 = SVC(kernel='rbf', gamma='auto')
svm_r1.fit(X_train, y_train)
```

```python
svm_p_test_labels = svm_p.predict(X_test)
svm_p0_test_labels = svm_p0.predict(X_test)
svm_p1_test_labels = svm_p1.predict(X_test)
svm_r_test_labels = svm_r.predict(X_test)
svm_r1_test_labels = svm_r1.predict(X_test)
```

```python
print(svm_p_test_labels)
print(y_test, len(y_test))
count_pos = len(y_test[np.where(y_test == 1)])
count_neg = len(np.where(y_test == 0)[0])
print(count_pos, count_neg)
```

### evaluation of the models

```python

def evaluation(predict_labels, true_labels):
    TP = 0
    TN = 0
    FP = 0
    FN = 0

    countAll = 0
    for i in range(len(y_test)):
        pred_label = predict_labels[i]
        true_label = true_labels[i]
        if (pred_label == true_label) and pred_label == 0: # i.e. true_lable = 0 NEG
            TN += 1
        elif (pred_label == true_label) and pred_label == 1: # i.e. true_lable = 1 POS
            TP += 1
        elif (pred_label != true_label) and pred_label == 0: # i.e. true_lable = 1 POS
            FN += 1
        elif (pred_label != true_label) and pred_label == 1: # i.e. true_lable = 0 NEG
            FP += 1
        countAll += 1


    print('true_pos_rate: ', TP / (TP + FN))
    print('true_neg_rate: ', TN /(TN + FP))
    print('false_pos_rate: ', FP / (FP + TN))
    print('false_neg_rate: ', FN /(TP + FN))
    print('accuracy: ', (TP + TN) / (TP + TN + FP + FN))
    print('precision: ', TP / (TP + FP)) # out of all predicted positive labels, how many were truely positive
    print('recall: ', TP / (TP + FN)) # out of all true positive labels in testset, how many were predicted right
    print('out of all true negative labels, correct-prediction rate: ', TN / (TN + FP))

print('---------------------------')
print('svm_p model: ')
evaluation(svm_p_test_labels, y_test)
print('---------------------------')
print('svm_p0 model: ')
evaluation(svm_p0_test_labels, y_test)
print('---------------------------')
print('svm_p1 model: ')
evaluation(svm_p1_test_labels, y_test)
print('---------------------------')
print('svm_r model: ')
evaluation(svm_r_test_labels, y_test)
print('---------------------------')
print('svm_r1 model: ')
evaluation(svm_r1_test_labels, y_test)
```
